**Namespaces**
- A namespace is a space that holds name(identifiers). Programmatically speaking, namespaces are dictionary of identifiers(keys) and their objects(values)

There are 4 Tyes of Namespaces
- Builtin Namespaces
- Global Namespaces
- Enclosing Namespaces
- Local Namespaces

**Scope and LEGB Rule**
- A scope is textual region of a python program where a namespace is directly accessible
- The interpreter searches for a name from the inside out looking in the local, enclosing, global, and finally the buil-in scope. If the  interpreter dosen't find the name in any of these locations, then python raises a NameError exception

````python
## Local and Global
a = 2 # Global Scope

def temp():
    b = 3 # Local Scope
    print(b)

temp()
print(a)
````

````python
## Local and Global Same Name

a = 2

def temp():
    a = 3
    print(a)

temp()
print(a)
````

````python
## Local dose not have but global has
a = 2

def temp():
    print(a)

temp()
print(a)
````

````python
a = 2

def temp():
    a += 1 ## We can access global varibale but we cant change the variable or update variable
    print(a)

temp()
print(a)
````

````python
a = 2

def temp():
    global a
    a += 1
    print(a)

temp()
print(a)
````

````python
def temp():
    global a
    a = 1
    print(a)

temp()
print(a)
````

#### Decorators
A Decorators in python is a function that recives another function as input and adds some functionality(decoration) to and it and returns it

This can happen only because python functons are 1st classcitizens

There are 2 types of decorators available in python
- Built in decorators like ```@staticmethod```,```@classmethod```,```@abstractmethod``` and ```@property``` etc
- User Defined Decorators that we programmers can create according to our needs

````python
## Python are 1st class function
def func():
    print('hello')
a = func()
a
````

````python
## Sample example
def dec(func):
    def wrapper():
        print('********************')
        func()
        print('********************')
    return wrapper
@dec
def greet():
    print('hello how are you')

greet()

@dec
def display():
    print('Hello Kishore')

display()
````

````python
## Anything meaningful
import time

def timer(func):
    def wrapper():
        start = time.time() # current time taken
        func() ## This code will work with one kind of function that is with no arguments
        print('time taken by ',func.__name__,time.time()-start,'secs')
    return wrapper
@timer
def greet():
    print("Good Morning")
    time.sleep(3)

@timer
def hello():
    print('hello wrold')
    time.sleep(3)

greet()
hello()
````

````python
@timer
def square(num):
    time.sleep(1)
    return num**2

square(3)
````

````python
import time

def timer(func):
    def wrapper(*args):
        start = time.time() # current time taken
        func(*args) ## This code will work with one kind of function that is with no arguments
        print('time taken by ',func.__name__,time.time()-start,'secs')
    return wrapper
````

````python
@timer
def square(num):
    time.sleep(1)
    print(num**2)

square(3)
````

````python
@timer
def power(a,b):
    print(a**b)
    time.sleep(3)

power(2,3)
````

````python
class InvalidTypeError(Exception):
    pass

def checkdatatype(datatype):
    def outer_wrapper(func):
        def inner_wrapper(*args):
            if type(*args) == datatype:
                return func(*args)
            else:
                raise InvalidTypeError(f"Expected {datatype.__name__}, got {type(args[0]).__name__}")
        return inner_wrapper
    return outer_wrapper
````

````python
@checkdatatype(int)
def square(num):
    return num * num

print(square(5))       
print(square("5"))     
````

````python
@checkdatatype(str)
def greet(name):
    print(f'hello {name}')

greet('kishore')
````
