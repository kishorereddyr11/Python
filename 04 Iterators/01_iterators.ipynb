**What is an Iteration**

Iteration is a general term for taking each item of something, one after another. Any time you use a loop, explicit or implicit, to go over a group of items, that is iteration

```python
lst = [1,2,3,4]

for num in lst:
    print(num)
```

**What is an Iterator**

An Iterator is an object that allows the programmer to traverse through a sequence of data without having to store entire data in the memory

```python
import sys

lst = [x for x in range(1,100000)]

print(sys.getsizeof(lst)/1024)

r = range(1,10000000000)

print(sys.getsizeof(r)/1024)
```

**What is Iterable**

Iterable is an object, which one can iterate over

It generates an iterator when passed to iter() method

```python
lst = [1,2,3]

print(type(lst))

print(type(iter(lst)))
```

#### **Points to Remember**

- Every **iterator** is **iterable**
- Not all **Iterables** are **iterators**

```python
lst = [1,2,3]

dir(lst) # __iter__ function is available then it is Iterable
```
