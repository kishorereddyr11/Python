#### Exceptions
if things go wrong during the exception of the program(runtime). it generally happens when something unforseen has happened 
- Exceptions are raised by python
- You have to take is on the fly

***Examples***
- Memory Overflow
- Divide by 0 -> logical Error
- DataBase Error

```python
## lets create a file
with open('sample.txt','w') as file:
    file.write('Hello World')
```

```python
## Try Catch Demo
try:
    with open('sample1.txt','r') as file:
        print(file.read())
except:
    print('sorry file not found')
```

```python
## Catching specific exception
try :
    m=10
    file = open('sample.txt','r')
    print(file.read())
    print(m)
    print(5/2)
    lst = [1,3,2]
    lst[23]
except FileNotFoundError:
    print("File Not Found Error")
except NameError:
    print("Variable Not Found Error")
except ZeroDivisionError:
    print("Can't Divide by zero")
except Exception as e: #Generic Except Statment, Always at last
    print(e)
```

```python
## else
try:
    file = open('sample.txt','r')
except FileNotFoundError:
    print("File not found")
except Exception as e:
    print(e)
else: ## if except triggered else will not execute
    print(file.read())
```

```python
## finally
try:
    file = open('sample.txt','r')
except FileNotFoundError:
    print("File not found")
except Exception as e:
    print(e)
else: ## if except triggered else will not execute
    print(file.read())
finally:
    print('Finally is always Executed')
```

```python
class Bank:

    def __init__(self,balance):
        self.balance = balance
    
    def withdraw(self,amount):
        if self.balance < amount:
            raise Exception('No Enough Balance')
        if amount < 0:
            raise Exception('Amount can not be -ve ')
        self.balance = self.balance - amount

obj = Bank(1000)
try :
    obj.withdraw(5000)
except Exception as e:
    print(e)
else:
    print(obj.balance)
```

```python
## Custom Exception Class
class MyException(Exception):
    def __init__(self,message):
        print(message)

class Bank:

    def __init__(self,balance):
        self.balance = balance
    
    def withdraw(self,amount):
        if self.balance < amount:
            raise MyException('No Enough Balance')
        if amount < 0:
            raise MyException('Amount can not be -ve ')
        self.balance = self.balance - amount

obj = Bank(1000)
try :
    obj.withdraw(5000)
except MyException as e:
    pass
else:
    print(obj.balance)
```
