**Banking Application**

```python
import json
import os
## Banking Application
class ATM:

    # Constructor
    def __init__(self):
        self.data_file = 'atm_data.json'
        self.pin = ''
        self.balance = 0
        self.load_data()
        self.menue()

    # Load the data from json file
    def load_data(self):
        # if file exists, load data
        if os.path.exists(self.data_file):
            with open(self.data_file, 'r') as file:
                data = json.load(file)
                self.pin = data.get('pin', '')
                self.balance = data.get('balance', 0)
        else:
            # create file automatically if missing
            with open(self.data_file, 'w') as file:
                json.dump({'pin': '', 'balance': 0}, file)
            print("Data file created successfully!")

    # Save data as json file
    def save_data(self):
        with open(self.data_file, 'w') as file:
            json.dump({'pin': self.pin, 'balance': self.balance}, file)

    # Menue Function
    def menue(self):
        user_input = input('''
        Hi How Can I Help You?
        1. Press 1 to create pin
        2. Press 2 to change pin
        3. press 3 to check balance
        4. press 4 to withdraw
        5. Exit
        ''')
        print(f'You Entered {user_input}')

        if user_input == '1':
            self.create_pin()
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            self.check_balance()
        elif user_input == '4':
            self.withdraw()
        elif user_input == '5':
            print("Thank you for using ATM ")
        else:
            print("Invalid option!")
            self.menue()
    
    # Create Pin
    def create_pin(self):
        if self.pin:
            print('Pin already exists! Use option 2 to change it.')
            return
        
        user_pin = input("Enter your pin : ")
        self.pin = user_pin

        user_balance = int(input("Enter your balance : "))
        self.balance = user_balance
        self.save_data()

        print('Pin Created Successfully')
        self.menue()

    # Change Pin
    def change_pin(self):
        user_old_pin = input('Enter your old pin : ')
        
        if user_old_pin == self.pin:
            # let him change the pin
            user_new_pin = input('Enter new pin : ')
            self.pin = user_new_pin
            self.save_data()
            print("Pin Changed Successfully!")
            self.menue()
        else:
            print("Wrong Pin")
            self.menue()
    
    # Check Balance
    def check_balance(self):
        user_pin = input('Enter your pin : ')
        
        if user_pin == self.pin:
            print('Your balance is : ', self.balance)
        else:
            print('Wrong Pin number')
        self.menue()

    # Withdraw Balance
    def withdraw(self):
        user_pin = input('Enter your pin : ')
        
        if user_pin == self.pin:
            amount = int(input('Enter amount for withdraw : '))
            if amount <= self.balance:
                self.balance -= amount
                self.save_data()
                print('Withdraw Successful')
                print('Current balance is : ', self.balance)
            else:
                print("Insufficient balance")
        else:
            print('Wrong Pin ')
        self.menue()



```

**Golden Rule of OOPS**

The **members** and **methods** in a class can be accessed by **only**  **object of the class itself**.

```python
class Temp:
    def __init__(self):
        print('address of self : ',id(self))
        self.name = 'Kishore'
```

```python
obj = Temp()
print('address of object : ',id(obj))

## Self is object of the Temp class only they are pointed to same address
```

**Custom DataType Fraction**

```python
class Fraction:

    def __init__(self,x,y):
        self.x = x
        self.y = y

    def __str__(self):
        return f'{self.x}/{self.y}'
    
    def __add__(self,other):
        new_x = self.x*other.y + other.x*self.y
        new_y = self.y*other.y

        return f"{new_x}/{new_y}"
    
    def __sub__(self,other):
        new_x = self.x*other.y - other.x*self.y
        new_y = self.y*other.y

        return f"{new_x}/{new_y}"

    def __mul__(self,other):
        new_x = self.x * other.x
        new_y = self.y*other.y

        return f"{new_x}/{new_y}"
    

    def __truediv__(self,other):
        new_x = self.x * other.y
        new_y = self.y*other.x

        return f"{new_x}/{new_y}"
    
    def convert_to_decimal(self):
        return self.x/self.y


    
```

```python
obj1 = Fraction(3,4)
obj2 = Fraction(1,2)
print(obj1 + obj2)
print(obj1 - obj2)
print(obj1 * obj2)
print(obj1 / obj2)

print(obj1.convert_to_decimal())
```
