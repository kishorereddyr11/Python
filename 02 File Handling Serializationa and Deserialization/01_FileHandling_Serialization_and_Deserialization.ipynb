````python
## readline -> read line by line , Readline will change line , print will change line
file = open('sample1.txt','r')
print(file.readline(),end ="")
print(file.readline())
````

````python
## When to use read and when to use readline
## if the file have lot of lines you do not want to load all the txt data into memory then you will use readline
## if the file have very less lines you use read function
## How would you know a file have this many lines
file = open("sample1.txt",'r')
while True:
    data = file.readline()
    if data == "":
        break
    else:
        print(data,end="")
file.close()
````

#### Using Context Manager (with)
- its good idea to close a file after usage as it will free up the resources
- if we do not close it, garbage collector would not close it
- with keyword closes the file as soon as the useage is over

````python
with open('sample.txt','w') as file:
    file.write("Iam Kishore Reddy")
````

````python
with open('sample1.txt','r') as file:
    s = file.read()
    print(s)
````

````python
## moving within a file -> 10 char then 10 char
with open('sample.txt','r') as file:
    print(file.read(5))
    print(file.read(3))
    print(file.read(5))
    print(file.read(4))
````

````python
## To load a big file in memory
big_lst = ['hello world' for i in range(1001)]

with open('sample3.txt','w') as file:
    file.writelines(big_lst)

with open('sample3.txt','r') as file:
    chunk_size = 11
    while True:
        s = file.read(chunk_size)
        print(s,end = "  ")
        if len(s) < chunk_size:
            break
````

````python
with open('sample.txt','r') as file:
    chunk_size = 3
    while True:
        s = file.read(chunk_size)
        print(s)
        if len(s) < chunk_size:
            break
````

````python
## Tell() function says. where the cursor is located
with open('sample.txt','r') as file:
    print(file.read(3))
    print(file.read(3))
    print(file.tell())
````

````python
## seek(), by using the seek function we can able to handle cursor
with open('sample.txt','r') as file:
    print(file.read(4))
    file.seek(0)## again the cursor is set to 0th index
    print(file.read(4))
````

````python
## Seek during write function
with open('sample.txt','w') as file:
    file.write('Hello World')
    file.seek(0)
    file.write('X')
````

````python
with open('sample.txt','a') as file:
    file.write(" Iam Kishore")
    file.seek(5)
    file.write("Hello ")
````

````python
## Working with Binary file
with open('image.png','rb') as file:
    with open('screenshot.png','wb') as writeFile:
        writeFile.write(file.read())
````

````python
## Working with other data type
with open('sample1.txt','w') as file:
    file.write('5')
````

#### Serialization and Deserialization
- serialization - process of converting data types to JSON formate
- Deserialization - process of converting JSON to python data types

````python
## Serialization using json module
## list
import json

lst = [1,2,3,4]
with open('demo.json','w') as file:
    json.dump(lst, file)
````

````python
d = {
    'name' : 'kishore',
    'age' : 22,
    'gender' : 'male'
}
with open('demo.json','w') as file:
    json.dump(d, file,indent=4)
````

````python
## Deserialization
with open('demo.json','r') as file:
    d = json.load(file)
    print(d)
    print(type(d))
````

````python
## SErialization and deserialize tuple
t = (1,2,3,4,5)
with open('demo.json','w') as file:
    json.dump(t,file) ## if we dump tuple it automatically convert to list we can not store tuple
````

#### Pickling
```pickling``` is the process where by a python object hierarchy is converted into a byte stream, and ```unpickling``` is the inverse operation, where by a byte stream (from binary file or bytes-like-object) is converted into object hierarchy

````python
class Person:

    def __init__(self,name, age):
        self.name = name
        self.age = age
    
    def display_info(self):
        print('Hii my name is ',self.name,' and I am ',self.age,' years old')
````

````python
person = Person('Kishore',22)
````

````python
import pickle

with open('person.pkl','wb') as file:
    pickle.dump(person,file)
````

````python
with open('person.pkl','rb') as file:
    p = pickle.load(file)
    p.display_info()
````
